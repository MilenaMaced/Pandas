## Lidando com valores ausentes

Como lidar com valores ausentes?
1. descatar;
2. substituir.

Colunas inteiras devem ser descartadas apenas se a maioria das entradas na coluna estiverem vazias.
Em nosso conjunto de dados, nenhuma das colunas está vazia o suficiente para sair completamente. Temos alguma liberdade para escolher qual método substituir os dados; no entanto, alguns métodos podem parecer mais razoáveis ​​que outros. Aplicaremos  métodos em algumas colunas diferentes:

1. Substitua por média:

* "normalized-losses": 41 dados ausentes
* "stroke": 4 dados ausentes
* "bore": 4 dados ausentes
* "horsepower": 2 missing data, replace them with mean
* "peak-rpm": 2 dados ausentes

2. Substitua pela frequência:

* "num-of-doors": 2 dados ausentes, substitua-os por "quatro"
  * Razão: 84% dos sedãs são de quatro portas. Como as quatro portas são mais frequentes, é mais provável que ocorra.

3. Apagar a linha inteira:

* "price": 4 dados ausentes, basta excluir a linha inteira
    * Razão: preço é o que queremos prever. Qualquer entrada de dados sem dados de preços não pode ser usada para previsão.



Calcular a média das colunas do tópico 1, transformando as mesmas em float, pois no DataFrame está como object, para posteriormente calcular a média:

```
med_normalized_losses = df['normalized-losses'].astype('float').mean(axis=0)
med_stroke = df['stroke'].astype('float').mean(axis=0)
med_bore = df['bore'].astype('float').mean(axis=0)
med_horsepower= df['horsepower'].astype('float').mean(axis=0)
med_peak_rpm = df['peak-rpm'].astype('float').mean(axis=0)
```

Substituindo "NaN" pelo valor médio em todas as colunas:

```
df['normalized-losses'].replace(np.nan,med_normalized_losses,inplace=True)
df['stroke'].replace(np.nan,med_stroke,inplace=True)
df['bore'].replace(np.nan,med_bore,inplace=True)
df['horsepower'].replace(np.nan,med_horsepower,inplace=True)
df['peak-rpm'].replace(np.nan,med_peak_rpm,inplace=True)
```

# Formatando os dados

Etapa crucial é alterar os formatos dos dados, para isso vamos utilizar o método <code>.dtype()</code> para verificar qual são os tipos dos dados e para alterarmos utilizamos o metódo <code>.astype()</code>.

```
df.dtypes
```

Como podemos ver acima, algumas colunas não são do tipo de dados correto. Variáveis ​​numéricas devem ter o tipo 'float' ou 'int', e variáveis ​​com cadeias de caracteres como categorias devem ter o tipo 'objeto'. Por exemplo, variáveis ​​'bore' e 'stroke' são valores numéricos que descrevem os motores; portanto, devemos esperar que sejam do tipo 'float' ou 'int'; no entanto, eles são mostrados como o tipo 'objeto'. 

Convertendo alguns dados para seus tipos apropriados

```
df[["bore", "stroke"]] = df[["bore", "stroke"]].astype("float")

df[["normalized-losses"]] = df[["normalized-losses"]].astype("int")

df[["price"]] = df[["price"]].astype("float")

df[["peak-rpm"]] = df[["peak-rpm"]].astype("float")

df["horsepower"]=df["horsepower"].astype(int, copy=True)
```

Finalmente obtemos o conjunto de dados limpo sem valores ausentes e todos os dados em seu formato apropriado.

# Padronização dos Dados

Os dados  são coletados em diferentes lugares com diferentes formatos.

*  que é padronização?

  Padronização é o processo de transformar dados em um formato comum que permite ao pesquisador fazer uma comparação significativa.

* Exemplo:

  **Transforme mpg em L / 100km:**

  Em nosso conjunto de dados, as colunas de consumo de combustível "city-mpg" e "highway-mpg" são representadas pela unidade de mpg (milhas por galão). Suponha que estamos desenvolvendo uma aplicação em um país que aceita o consumo de combustível com o padrão L / 100km

  Precisamos aplicar a transformação de dados para transformar mpg em L / 100km?

**A fórmula para a conversão de unidades é**

L / 100km = 235 / mpg

```
df['city-L/100km'] = 235/df["city-mpg"]
df.head()
```

Transformando mpg em L / 100km na coluna "highway-mpg" e alterando o nome da coluna para "highway-L / 100km".

```
df["highway-mpg"] = 235/df["highway-mpg"]
df.rename(columns={'highway-mpg':'highway-L/100km'}, inplace=True)
```

# Normalização de Dados

Normalização é o processo de transformar valores de várias variáveis ​​em um intervalo semelhante. As normalizações típicas incluem escalar a variável para que a média da variável seja 0, escalar a variável para que a variação seja 1 ou escalar variável para que os valores da variável variem de 0 a 1.

As três técnicas normalização mais utilizadas

Escalar Simples:

<h2> $ x_n = \frac {x_{value}} {x_{max}}$ </h2>

Min-Max:
<h2>$ x_n = \frac {x_{value}-x_{min}} {x_{max}-x_{min}}$</h2>

Escore Padrão:

<h2> $x_n = \frac {x_{value}-\mu} {\sigma}$ </h2>


$\mu$=  média **(.mean())**

$\sigma$ = Desvio Padrão **(.std())**

Para demonstrar a normalização, digamos que queremos dimensionar as colunas "length", "width" e "height"

Gostaria de Normalizar essas variáveis ​​para que seus valores variem de 0 a 1, utilizando a formúla Escala simples.

```
df['length'] = df['length']/df['length'].max()
df['width'] = df['width']/df['width'].max()
df['height'] = df['height']/df['height'].max()
```

# Binning

Por que binning?

Binning é um processo de transformar variáveis ​​numéricas contínuas em 'categorias' discretas, para análise agrupada.

Exemplo:

Em nosso conjunto de dados, "horsepower" é uma variável com valor real, variando de 48 a 288, e possui 57 valores únicos. E se nos importarmos apenas com a diferença de preço entre carros com alta potência, potência média e pouca potência (3 tipos)? Podemos reorganizá-los em três 'categorias' para simplificar a análise?



Usaremos a função numpy para obter o intervalo para poder categorizar :

 <code>np.linspace(start_value, end_value, numbers_generated)</code>.

* Como queremos incluir o valor mínimo de potência, queremos definir start_value = min (df ["horsepower"]).

* Como queremos incluir o valor máximo de potência, queremos definir end_value = max (df ["horsepower"]).

* Como estamos construindo 3 posições de igual comprimento, deve haver 4 divisores, então numbers_generated = 4.

```
bins = np.linspace(min(df["horsepower"]), max(df["horsepower"]), 4)
bins
```

Os grupos são 'Low', 'Medium' e 'High'

```
nome_grupos= ['Low', 'Medium', 'High']
```

Aplicamos a função "cut" para determinar a que cada valor de "df ['horsepower']" pertence. Segmenta e classificar os valores dos dados em compartimentos.

```
df['horsepower-binned'] = pd.cut(df['horsepower'], bins, labels=nome_grupos, include_lowest=True )
df[['horsepower','horsepower-binned']].head(20)
```

```
df['horsepower-binned'].value_counts()
```

# Variável de indicador

O que é uma variável indicadora?

Uma variável indicadora (ou variável dummy) é uma variável numérica usada para rotular categorias. Eles são chamados de 'manequins' porque os próprios números não têm significado inerente.

Por que usamos variáveis ​​indicadoras?

Portanto, podemos usar variáveis ​​categóricas para análise de regressão nos módulos posteriores.

Exemplo:

Vemos que a coluna "tipo de combustível" tem dois valores únicos, "gás" ou "diesel". Regressão não entende palavras, apenas números. Para usar esse atributo na análise de regressão, convertemos "tipo de combustível" em variáveis ​​indicadoras. Igual o exemplo abaixo:



 Tipo de combustível | Gás  | Diesel  |
 :--:                | :--: |   :--:  |
Gás                  | 1    | 0       |
Diesel               | 0    | 1       |
Gás                  | 1    | 0       |

Usaremos o método do panda 'get_dummies' para atribuir valores numéricos a diferentes categorias de tipo de combustível.

```
dummy_variable = pd.get_dummies(df["fuel-type"])
dummy_variable.head()
```

Agora temos o valor 0 para representar "gás" e 1 para representar "diesel" na coluna "tipo de combustível". Agora vamos inserir esta coluna novamente em nosso conjunto de dados original.

```
df = pd.concat([df, dummy_variable], axis=1)
df.drop("fuel-type", axis = 1, inplace=True)
```

Criando uma variável indicadora para a coluna "aspiration": "std" para 0, enquanto "turbo" para 1.

```
dummy_variable_2 = pd.get_dummies(df['aspiration'])
dummy_variable_2.rename(columns={'std':'aspiration-std', 'turbo': 'aspiration-turbo'}, inplace=True)
dummy_variable_2.head()
```

Mesclar o novo quadro de dados ao quadro de dados original e soltar a coluna 'aspiration'

```
df = pd.concat([df, dummy_variable_2], axis=1)
df.drop('aspiration', axis = 1, inplace=True)
```
